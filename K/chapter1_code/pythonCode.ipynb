# how to plot vectors in python
## first method

```python
import numpy as np
import matplotlib.pyplot as plt
V = np.array([[1,1],[-2,2],[4,-7]])
origin = np.array([[0,0,0],[0,0,0]])#origion point
plt.quiver(*origin,V[:,0],V[:,1],color = ["r","b","g"],scale =21)
plt.grid()
plt.show()
```

# second method


```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
soa = np.array([[0, 0, 1, 1, -2, 0], [0, 0, 2, 1, 1, 0],
                [0, 0, 3, 2, 1, 0], [0, 0, 4, 0.5, 0.7, 0]])
X,Y,Z,U,V,W = zip(*soa)
fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")
ax.quiver(X,Y,Z,U,V,W)
ax.set_xlim([-1,0.5])
ax.set_ylim([-1,1.5])
ax.set_zlim([-1,8])
plt.show()
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection ="3d")
x=[0,1,2,3,4,5,6]
y=[0,1,4,9,16,25,36]
z=[0,1,4,9,16,25,36]
ax.plot3D(x,y,z,"red")
ax.scatter3D(x,y,z,c=z,cmap="cividis");
plt.show()
```

# vector fields

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))
u = 1
v = -1
plt.quiver(x,y,u,v)
plt.grid()
plt.show
```

```python
x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

u = x/np.sqrt(x**2 + y**2)
v = y/np.sqrt(x**2 + y**2)

plt.quiver(x,y,u,v)
plt.grid()
plt.show()
```
